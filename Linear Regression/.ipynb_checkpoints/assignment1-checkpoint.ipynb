```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error , r2_score
from pandas.core.common import random_state
```

```python
df_crp = pd.read_csv('canada_per_capita_income.csv')
```

```python
df_crp.head()
```

```python
df_crp.tail()
```

```python
df_crp.info()
```

```python
plt.title('year vs income')
plt.xlabel('years')
plt.ylabel('income')
plt.scatter(df_crp['year'] , df_crp['per capita income (US$)'])
plt.box(False)
plt.grid(True)
plt.show()
```

```python
X = df_crp.iloc[: ,:1] 
```

```python
X
```

```python
 y = df_crp.iloc[:,1:]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_test_split(X , y , test_size = 0.2)
```

```python
X_train, X_test , y_train, y_test = train_test_split(X , y , test_size = 0.2, random_state = 0)
```

```python
X_train, X_test , y_train, y_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred_test = regressor.predict(X_test)
```

```python
y_pred_test 
```

```python
y_pred_train = regressor.predict(X_train)
```

```python
y_pred_train
```

```python
print(regressor.intercept_)    #intercept = c

print(regressor.coef_)        #coefficient = m
```

```python
# evaluate the model
print('MSE = ', mean_squared_error(y_test, y_pred_test ))
print('R2 = ', r2_score(y_test, y_pred_test ))
print('RMSE = ', np.sqrt(mean_squared_error(y_test, y_pred_test )))
```

```python
plt.scatter(X_test, y_test, color ='red')
plt.plot(X_test, y_pred_test,marker = 'o' ,color ='lightcoral')
plt.title('year vs income (Test Set)')
plt.xlabel('Years ')
plt.ylabel('income')
plt.legend(['X_test/y_test', 'X_test/y_pred_test'], title = 'year/income', loc='best', facecolor='white')
plt.box(False)
plt.show()
```
