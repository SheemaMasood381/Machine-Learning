## Encoding gender column

```python
from sklearn.preprocessing import LabelEncoder 


enc = LabelEncoder()
cus_df.Gender = enc.fit_transform(cus_df.Gender)
```

```python
cus_df.sample(5)
```

## Perform Kmeans Clustering:

```python
k = 5

kmeans = KMeans(n_clusters = k, random_state = 42)
kmeans.fit(cus_df)
```

```python
# Get cluster labels
labels = kmeans.labels_

# Add the cluster labels to your DataFrame
cus_df['Cluster'] = labels
```

```python
cus_df.sample(5)
```

```python
# Determine the optimal number of clusters using the Elbow Method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(cus_df)
    wcss.append(kmeans.inertia_)

# Plot the Elbow curve
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

### DIMENTIONALITY REDUCTION

```python
from sklearn.decomposition import PCA

# Reduce dimensions to 2D for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(cus_df.drop('Cluster', axis=1))
```

```python
# Create a DataFrame for plotting
pca_df = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
pca_df['Cluster'] = labels

# Plot the clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='Cluster', palette='viridis', s=100, alpha=0.7)
plt.title('Clusters Visualization')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# Compute Silhouette Score
silhouette_avg = silhouette_score(cus_df, labels)
print(f'Silhouette Score: {silhouette_avg:.3f}')
```
